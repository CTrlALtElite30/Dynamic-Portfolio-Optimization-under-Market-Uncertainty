```python
# --- Setup & imports ---
import os, sys
import numpy as np
import pandas as pd

PROJECT_ROOT = r"E:/project"
if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)

EVAL_PATH = os.path.join(PROJECT_ROOT, "outputs", "evaluations")

BASELINE_CSV = os.path.join(EVAL_PATH, "baseline_buy_and_hold_equity_test.csv")
DQN_CSV      = os.path.join(EVAL_PATH, "dqn_discrete_equity_test.csv")
PPO_CSV      = os.path.join(EVAL_PATH, "ppo_continuous_equity_test.csv")

def load_equity_csv(path):
    if not os.path.exists(path):
        raise FileNotFoundError(f"Missing file: {path}")
    df = pd.read_csv(path)
    # Try standard 'equity' column; otherwise fall back to the first column
    if "equity" in df.columns:
        eq = df["equity"].astype(float).values
    else:
        # handle unnamed or different column names (e.g., '0')
        eq = df.iloc[:, 0].astype(float).values
    return np.asarray(eq, dtype=float)

# --- Load the three equity curves ---
baseline_eq = load_equity_csv(BASELINE_CSV)
dqn_eq = load_equity_csv(DQN_CSV)
ppo_eq = load_equity_csv(PPO_CSV)

print("Loaded equity lengths:")
print(f"  Baseline: {len(baseline_eq)}")
print(f"  DQN     : {len(dqn_eq)}")
print(f"  PPO     : {len(ppo_eq)}")

print("\nPreview first 5 values:")
print("  Baseline:", baseline_eq[:5])
print("  DQN     :", dqn_eq[:5])
print("  PPO     :", ppo_eq[:5])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(baseline_eq, label="Baseline (Buy & Hold)", linewidth=2)
plt.plot(dqn_eq, label="DQN Agent", linewidth=2)
plt.plot(ppo_eq, label="PPO Agent", linewidth=2)

plt.title("Equity Curve Comparison", fontsize=14)
plt.xlabel("Time Steps", fontsize=12)
plt.ylabel("Portfolio Value", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
def compute_drawdown(equity):
    cummax = np.maximum.accumulate(equity)
    drawdown = (equity - cummax) / cummax
    return drawdown

def compute_rolling_returns(equity, window=30):
    returns = pd.Series(equity).pct_change().fillna(0)
    rolling = returns.rolling(window).mean()
    return rolling

# --- Compute ---
baseline_dd = compute_drawdown(baseline_eq)
dqn_dd      = compute_drawdown(dqn_eq)
ppo_dd      = compute_drawdown(ppo_eq)

baseline_roll = compute_rolling_returns(baseline_eq)
dqn_roll      = compute_rolling_returns(dqn_eq)
ppo_roll      = compute_rolling_returns(ppo_eq)

# --- Plot Rolling Returns ---
plt.figure(figsize=(12,6))
plt.plot(baseline_roll, label="Baseline Rolling Return", alpha=0.7)
plt.plot(dqn_roll, label="DQN Rolling Return", alpha=0.7)
plt.plot(ppo_roll, label="PPO Rolling Return", alpha=0.7)
plt.title("Rolling Returns (30-step window)")
plt.xlabel("Time Steps")
plt.ylabel("Return")
plt.legend()
plt.grid(True)
plt.show()

# --- Plot Drawdowns ---
plt.figure(figsize=(12,6))
plt.plot(baseline_dd, label="Baseline Drawdown", alpha=0.7)
plt.plot(dqn_dd, label="DQN Drawdown", alpha=0.7)
plt.plot(ppo_dd, label="PPO Drawdown", alpha=0.7)
plt.title("Drawdowns")
plt.xlabel("Time Steps")
plt.ylabel("Drawdown")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# --- Compute Cumulative Returns (%)
baseline_cum = (baseline_eq / baseline_eq[0] - 1) * 100
dqn_cum      = (dqn_eq / dqn_eq[0] - 1) * 100
ppo_cum      = (ppo_eq / ppo_eq[0] - 1) * 100

# --- Plot ---
plt.figure(figsize=(12,6))
plt.plot(baseline_cum, label="Baseline (Buy & Hold)")
plt.plot(dqn_cum, label="DQN Agent")
plt.plot(ppo_cum, label="PPO Agent")
plt.title("Cumulative Returns (%)")
plt.xlabel("Time Steps")
plt.ylabel("Cumulative Return (%)")
plt.legend()
plt.grid(True)
plt.show()

# --- Final Cumulative Returns Table ---
final_returns = {
    "Baseline": baseline_cum[-1],
    "DQN": dqn_cum[-1],
    "PPO": ppo_cum[-1]
}

final_returns_df = pd.DataFrame(final_returns, index=["Final Cumulative Return (%)"]).T
display(final_returns_df.style.format("{:.2f}"))
```
