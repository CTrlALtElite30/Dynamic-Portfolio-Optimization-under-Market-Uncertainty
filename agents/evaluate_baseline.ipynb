```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Project Root
PROJECT_ROOT = "E:/project"
if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)

# Dataset
DATA_PATH = os.path.join(PROJECT_ROOT, "outputs", "datasets", "states_v1_single_asset.npz")
data = np.load(DATA_PATH)
prices = data["prices"]


# Baseline Evaluation (Buy & Hold)
def evaluate_buy_and_hold(prices, save_suffix="baseline"):
    initial_value = 100000
    equity_curve = initial_value * (prices / prices[0])

    # Compute metrics
    total_return = (equity_curve[-1] / equity_curve[0] - 1) * 100
    n_days = len(prices)
    CAGR = (equity_curve[-1] / equity_curve[0]) ** (252 / n_days) - 1
    returns = np.diff(equity_curve) / equity_curve[:-1]
    volatility = np.std(returns) * np.sqrt(252)
    sharpe = (CAGR) / (volatility + 1e-8)
    max_dd = np.min(equity_curve / np.maximum.accumulate(equity_curve) - 1) * 100
    calmar = CAGR / (abs(max_dd / 100) + 1e-8)
    downside = returns[returns < 0]
    sortino = CAGR / (np.std(downside) * np.sqrt(252) + 1e-8)

    metrics = {
        "total_return_pct": total_return,
        "CAGR_pct": CAGR * 100,
        "volatility_ann_pct": volatility * 100,
        "sharpe": sharpe,
        "max_drawdown_pct": max_dd,
        "calmar": calmar,
        "sortino": sortino,
        "n_days": n_days
    }

    # ---------- Save Outputs ----------
    eval_dir = os.path.join(PROJECT_ROOT, "outputs", "evaluations")
    plot_dir = os.path.join(PROJECT_ROOT, "outputs", "plots")
    os.makedirs(eval_dir, exist_ok=True)
    os.makedirs(plot_dir, exist_ok=True)

    # Save CSV
    pd.DataFrame({"equity": equity_curve}).to_csv(
        os.path.join(eval_dir, f"baseline_buy_and_hold_equity_{save_suffix}.csv"), index=False
    )

    # Save JSON
    import json
    with open(os.path.join(eval_dir, f"baseline_buy_and_hold_metrics_{save_suffix}.json"), "w") as f:
        json.dump(metrics, f, indent=4)

    # Save Plot
    plt.figure(figsize=(10,5))
    plt.plot(equity_curve, label="Buy & Hold")
    plt.title("Baseline: Buy & Hold")
    plt.xlabel("Days")
    plt.ylabel("Equity Value")
    plt.legend()
    plt.grid(True)
    plt.savefig(os.path.join(plot_dir, f"baseline_buy_and_hold_equity_{save_suffix}.png"))
    # plt.close()

    print(f"Baseline saved\nCSV: {eval_dir}\nJSON metrics: {eval_dir}\nPNG plot: {plot_dir}")
    return metrics

# # later on when all will work perfectly - utils - evaluation.py to save all records delete it if it causes error
# from utils.evaluation import evaluate_agent
# metrics_baseline = evaluate_agent(val_env.envs[0], model=None, name="baseline")

# Run baseline
metrics = evaluate_buy_and_hold(prices, save_suffix="test")
print("Metrics:", metrics)
```
